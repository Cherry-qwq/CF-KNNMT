```python
import matplotlib.pyplot as plt
import numpy as np

# 设置随机数种子以获得可重现的结果
np.random.seed(39)

# 生成95个红色点
red_points = np.random.rand(190, 2)  # 生成一个95x2的数组，每行代表一个点的x和y坐标

# 生成第一个蓝色点
blue_point = np.random.rand(1, 2)  # 生成一个1x2的数组，代表第一个蓝色点的x和y坐标

# 在第一个蓝色点的附近生成剩余的4个蓝色点
# 我们通过在第一个蓝色点周围添加一些小的随机偏移来实现这一点
offsets = np.random.normal(loc=0, scale=0.10, size=(9, 2))  # 生成4x2的数组，每行代表偏移量
blue_points = blue_point + offsets  # 在第一个蓝色点的基础上添加偏移量以生成剩余的蓝色点
blue_points = np.vstack([blue_point, blue_points])  # 将第一个蓝色点和其余蓝色点合并


# 设置图的大小
plt.figure(figsize=(4, 3))

top_blue_point = blue_points[blue_points[:, 1].argmax()]

# 设置圆的半径
radius = 0.1



circle_center = np.array([0.5, 0.5])
circle_radius = 0.5

# 过滤掉圆外的红色点
red_points_inside = np.array([point for point in red_points if np.linalg.norm(point - circle_center) < circle_radius])

# 过滤掉圆外的蓝色点
blue_points_inside = np.array([point for point in blue_points if np.linalg.norm(point - circle_center) < circle_radius])

# 找到从上往下数第二个蓝色点
sorted_blue_points = blue_points[np.argsort(-blue_points[:, 1])]  # 对蓝色点按y坐标降序排序
second_top_blue_point = sorted_blue_points[1]  # 从上往下数第二个蓝色点





# 重新绘制图表
plt.figure(figsize=(4, 4))
plt.scatter(red_points_inside[:, 0], red_points_inside[:, 1], color='green', s=10, alpha=0.6)
plt.scatter(blue_points_inside[:, 0], blue_points_inside[:, 1], color='blue', s=10, alpha=0.6)
plt.scatter(second_top_blue_point[0], second_top_blue_point[1], color='red', s=10, alpha=1)

# 添加以从上往下数第二个蓝色点为圆心的虚线圆
circle = plt.Circle(second_top_blue_point, radius, color='blue', fill=False, linestyle='--', linewidth=0.9)
plt.gca().add_artist(circle)



# 设置图表为圆形虚线边框
plt.gca().set_aspect('equal', adjustable='box')
plt.gca().add_patch(plt.Circle(circle_center, circle_radius, color='black', fill=False, linestyle='--', linewidth=1.5))

# 隐藏坐标轴刻度
plt.xticks([])
plt.yticks([])

# 隐藏坐标轴的边框线
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)

plt.show()

```
